IDEA 1:
Our database idea is for a dynamic airline ticket website. On this website, the user would have the universal ability to purchase plane tickets of different types. Specific ticket purchases could be made from any airline provider, as well as round trip or direct flight tickets (as long as they are available). While this may be compared to a travel agency website, it would have even more specific data about the flights, and not just the ticket. Some travel agency websites are limited in what information they provide to their consumers, however, our website would provide data such as the pilot background being made available to customers to allow them to make the most informed decision. This is an important part of the business because it would allow customers to become closer to the pilots and have a safer, more comfortable, and better feeling about the flight they are about to go on. This would also allow customers to vary flight providers, and decrease the potential of monopolies on the market, because they would read about pilots from different providers and give them the chance to choose a provider strictly based on the pilot. This business is being borrowed from a different business, as previously stated before, however would add more dynamic information to appeal to the customer. On a daily basis, the business would have to stay up to date with real time information on available flights and be able to transfer that information to consumers. Furthermore, customer data would be a necessity in having as they would have to display frequency and availability of flights as well as seats.  

END USER/REAL TIME TABLE:
The one end-user or real time table is the Pilot table. This is shown as real time because the data was collected through a survey from pilots to provide to consumers. The survey included the pilots first name, last name, their pilot number associated with them, and their years of experience piloting consumer planes. 

ENTITIES/ATTRIBUTES: 
The entities for this concept include pilot, airline information, flight information, ticket, and customer. The attributes for pilot include: first name, last name, pilot number, and years of experience. The attributes for airline information are airline provider, flight number, and pilot number. The attributes for flight information are flight number, takeoff time, takeoff location, and seats available. The attributes for ticket are type, class, flight number, and price. Finally, the attributes for customers are first name, last name, flight number, and takeoff time. Constraints would be given to takeoff time between 12:00AM - 11:59PM and ticket price would be anywhere between about $20 to $64,000. 

GOAL: 
Our goal is to figure out if there is correlation between pilot and flight efficency (such as takeoff time and price). These tables will help us answer that question because the attributes we chose directly provide information that can be linked together. 

IDEA 2:
This idea is a dynamic and updated version of an online library finder for students and professors at universities. While this is a common concept, this idea would be more specific and user-friendly for mobile users, as most of us are. Students and professors have limited time and finding textbooks and research materials and this is something we all dont want to be a hassle. Most online databases at universities are hard to use and do not provide enough information for the user to make informed decisions with. This online library concept is unique because it is an updated version of the current service universities have nationwide. It would provide more information on books, movies, authors, genres and finally locations. The mobile version of this database would show locations of books and items on an online map of the library. It would also be unique because it would allow the user to acquire their items faster than any other source, as the mapping would be up to date and show which books are stocked or checked out on a daily basis. A typical interaction through this dataset would be the user searching for the information they are looking for, then they would locate it on the map and star it for them to go find later. All of this data would be saved and backed up through the dataset, as it saves user information inputted. Finally, after the user has held this item in their possession the dataset will send a text or email to the user reminding them of their item being due. This allows for users to stay updated and the system to work flawlessly. 

END USER/REAL TIME TABLE:
The one end-user or real time table for this idea is the user table. Outside the library we would set up a table for students/professors who recently purchased a book and movie. They would fill out the last numbers of their book ID and the last numbers of their movie ID and we would associate an ID for them. This information would be inputted in the database and will continue to be inputted.

ENTITIES/ATTRIBUTES: (all tables are shown as one-to-one)
The entities include uses, employees, computers, books, and movies. Attributes for users include their ID, the last movie they checked out, and the last book they checked out. For employees attributes include ID, hours worked, first name, and last name. Computers attributes include ID, the last employee to access, and the last user to access. Attributes of books can be the ID number, the title, and the author. Movie attributes include Id number, director, and title. Constraints include an employee can only work 10 hours a day because that is as long as the library is open. 

GOAL: 
Our goal for this idea is to figure out which books are being purchased more by students or employee/professors. This will be done by taking the tables "user", "employees", and "books" to compare and contrast information. 

IDEA 3:
This is a database found on a music streaming mobile app. On this database, users can find music by artists that are in their area. The app will take the location of both users and artists for this feature. This will then allow the user to see artists that are in their area or got their start nearby. By having this feature, artists will get more exposure and users can discover new music. Users can navigate the app by the browsing through artists, genres, or the recommended feature of the app. If they browse by genre, the database will generate a list of new, popular songs of that genre. If they navigate through artists, they have to search up an artist themselves and then a list of similar artists will be generated. If they browse through the 'for you' section, they will see a list of songs based on the music they like. This idea is not original at all, it is very similar to many music streaming sites, like Spotify or Apple Music. It's different because it has a location feature and a review section but most music platforms have very similar ideas. The location feature is a new idea to help enhance user and artist experience. In the review section, users will be able to comment only after listening to the song. This allows the artist to  get real feedback and communicate with their fans. Fans get the opportunity to express their opinions and see what other people think. Users can create their own playlists and share them with other users as well. The database will be used daily for artists to upload music and users to stream it.

"END USER"/REAL TIME TABLE:
The popular song table collects data on how many streams a song gets. It then tells the user and artist the artists most popular song based on how many streams it has received. It is a real time example because as a song is streamed the most popular song is available to change. 

ENTITIES/ATTRIBUTES:
The entities for this concept include artists, artist location, most popular song, user, and user location. Attributes for artist may include first name, last name, ID number, age, and number of streams. The location of the artist includes home town state, home town city, current state, and current city. The most popular song includes artist ID, number of streams, and users that have streamed. Attributes for user include ID, first name, last name, hours streamed a day, and date registered. The location of the user includes hometown state, hometown city, current state, and current city. Constraints could be that number of streams cannot be less than zero.

GOAL: 
The goal for this idea is where the most popular artists are being streamed in the nation. We will answer this question by using the tables "user", "artist location" and "user location" to determine this relationship.

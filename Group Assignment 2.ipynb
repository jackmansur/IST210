To help the library keep track of its own data, an ER database with tables for information will be developed and used for reference. This diagram includes entities, attributes, PK’s, FK’s, relationships, connectivity, and cardinality. 

The most important table pertains to the entity of books. This table includes information (attributes) on book identification (PK), title, number available, and the employee that checked it out to a customer. The relationship between books and customers is zero to five because a customer does not necessarily have to rent any books, and they can only take out up to five at once, thus making the cardinality optional. The foreign key within this table is the employee ID, which leads into the next table (this cardinality between these tables being one to one either way). 

The employee table has a list of the employee identification (PK), first name, last name, and amount of hours per week. The identification connects it to the branch table for the branch that the employee works at. This relationship is many to one, specifically estimating 30 employees to one branch of the library.

The branch table includes information on branch identification (PK), the employees (FK), the location, and the books kept there. It can be connected to the bill table because of bill location. The relationship is one-to-many, specifically speaking in cardinality; there is one bill assigned from every branch and one branch sends out multiple bills, estimated at 1000. 

The bill table includes information on the identification (PK), location, customer, and the price of the bill. A customer may not rent a book and thus does not have to pay a bill, and since a customer can only rent five books at one time they only need to pay up to five bills, showing how this relationship is optional. Finally, The user table lists the first name, last name, user identification (PK), and the book identification of the book they checked out to help keep track of information and the cycle of the ER diagram.
